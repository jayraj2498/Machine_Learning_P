```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
centers = [[-1,-1] ,[2,2] ,[5,5]]
x , y = make_blobs(n_samples=600 , centers=centers)
```

```python
x
```

```python
set(y)
```

```python
colurset = np.array(['Red','blue','green','black'])
```

```python
plt.scatter(x[:,0],x[:,1], c=colurset[y])
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.5 , min_samples=6) 

db.fit(x) 

db_pred_dv =db.labels_
```

```python
set(db_pred_dv)   # -1 is the outlier 
```

```python
plt.scatter(x[:,0],x[:,1], c=colurset[db_pred_dv])   # all black pt are the outlier 
```

```python
# if you want to check accuary then 
```

```python
from sklearn.metrics import homogeneity_score
```

```python
homogeneity_score(y , db_pred_dv)
```
