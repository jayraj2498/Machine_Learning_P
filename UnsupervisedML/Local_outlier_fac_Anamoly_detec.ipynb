### By comparing the local density of a sample to the local densities of its neighbors, one can identify samples that have a substantially lower density than their neighbors. These are considered outliers.

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_circles 
from sklearn.neighbors import LocalOutlierFactor
```

```python
X , y =make_circles(n_samples=750 , factor=0.3 , noise=0.1 )  
```

```python
plt.scatter(X[ : , 0] , X[ : , 1] )
```

```python
X, y = make_circles(n_samples=750, factor=0.3, noise=0.1, random_state=42)
```

```python
# Create LocalOutlierFactor model
lof_model = LocalOutlierFactor(contamination=0.1)  # Adjust contamination based on your dataset
```

```python
y_pred = lof_model.fit_predict(X)
```

```python
y_pred
```

```python
plt.figure(figsize=(8, 8))


plt.scatter(X[y_pred == 1][:, 0], X[y_pred == 1][:, 1], label='Inliers', edgecolors='k', s=50)

plt.scatter(X[y_pred == -1][:, 0], X[y_pred == -1][:, 1], label='Outliers', edgecolors='r', s=50)

plt.title("Local Outlier Factor (LOF)")
plt.legend()
plt.show()
```
