## apply corelation 

```python
corr_matrix = x.corr()
```

```python
corr_matrix
```

```python
corr_matrix.columns
```

```python
corr_names_col = set()
```

```python
for row in range(len(corr_matrix)):
    for col in range(row) :
        if abs(corr_matrix.iloc[row][col]) > 0.2 :   # iloc bcoz we using dataframe 
            corr_names_col.add(corr_matrix.columns[row])
```
