```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature
```

```python
def correlation(dataset , threshold):
    col_corr_set= set()
    col_corr_lst = [] 
    
    print(f"set initiallize is {col_corr_set}")
    print(f"set initiallize is {col_corr_lst}")
    
    corr_arr = dataset.corr()  #  correation on x_trian matrix  
    
    for row in range(len(corr_arr)):
        for col in range(row) : 
            if abs(corr_arr.iloc[row ,col]) > threshold :          # why iloc bcoz we are using dataframe   
                colname = corr_arr.columns[row]                     # getting the name of column  
                
                col_corr_lst.append(colname)
                col_corr_set.add(colname)  
                
                print(f" print name which correalated : {colname} \n")
                print(f" list is :{col_corr_lst}")
                print(f" sets is : {col_corr_set}") 
                
                

    print(f" col of sets is  {col_corr_set}")
    
    return col_corr_set                              # we are sacing the data interm of sets bcoz it dont allow repetation 
    
```
